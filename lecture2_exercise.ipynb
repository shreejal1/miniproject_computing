```python
import psutil
import cpuinfo
import platform

def get_server_specs():
    # 1. CPU Core Counts
    physical_cores = psutil.cpu_count(logical=False)
    total_cores = psutil.cpu_count(logical=True)

    # 2. Memory (RAM)
    raw_mem = psutil.virtual_memory().total
    main_memory_gb = round(raw_mem / (1024 ** 3), 2)

    # 3. CPU Details
    info = cpuinfo.get_cpu_info()
    cpu_model = info.get('brand_raw', 'Unknown CPU')
    
    # On Linux, these might be integers or strings
    l1 = info.get('l1_instruction_cache_size', 'N/A')
    l2 = info.get('l2_cache_size', 'N/A')
    l3 = info.get('l3_cache_size', 'N/A')

    print(f"--- Hardware Specifications ({platform.system()}) ---")
    print(f"Hostname:       {platform.node()}")
    print(f"CPU Model:      {cpu_model}")
    print(f"Physical Cores: {physical_cores}")
    print(f"Logical Cores:  {total_cores}")
    print(f"Main Memory:    {main_memory_gb} GB")
    print(f"L3 Cache Size:  {l3}")

if __name__ == "__main__":
    get_server_specs()
```

```python
import os

def get_linux_cache():
    print(f"{'Cache Level':<15} | {'Size':<10} | {'Type':<10}")
    print("-" * 40)
    
    base_path = "/sys/devices/system/cpu/cpu0/cache/"
    
    if not os.path.exists(base_path):
        print("Could not find cache information in /sys/devices/system/cpu/")
        return

    # Iterate through indices (index0, index1, etc.)
    for index in sorted(os.listdir(base_path)):
        path = os.path.join(base_path, index)
        
        try:
            with open(os.path.join(path, "level"), "r") as f:
                level = f.read().strip()
            with open(os.path.join(path, "size"), "r") as f:
                size = f.read().strip()
            with open(os.path.join(path, "type"), "r") as f:
                cache_type = f.read().strip()
            
            level_name = f"L{level} Cache"
            print(f"{level_name:<15} | {size:<10} | {cache_type:<10}")
        except Exception:
            continue

if __name__ == "__main__":
    get_linux_cache()
```
